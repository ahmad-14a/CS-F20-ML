```python
import numpy as np
from PIL import Image
from sklearn.cluster import KMeans
from IPython.display import display, HTML

# Read and preprocess the image
image = Image.open('ironMan.jpg')
image_data = np.array(image)
reshaped_data = image_data.reshape(-1, 3)  # Reshape to 2D array

# Apply k-means clustering
k = 10  # Number of clusters
kmeans = KMeans(n_clusters=k, random_state=0)
kmeans.fit(reshaped_data)

# Get the labels and cluster centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Replace pixel values with centroid values
new_image_data = centroids[labels].reshape(image_data.shape)

# Convert the data back to image format
new_image = Image.fromarray(new_image_data.astype(np.uint8))

# Display the original and clustered images
#image.show()
#new_image.show()

display(image)
display(new_image)
```
