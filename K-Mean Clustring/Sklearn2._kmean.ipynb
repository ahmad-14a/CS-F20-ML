```python
import numpy as np
from PIL import Image
from sklearn.cluster import KMeans
from IPython.display import display, HTML

# Read and preprocess the image
image = Image.open('ironMan.jpg')

# Define the scaling factor for image size
scaling_factor = 1  # Adjust this value to change the size of the shrunk images

# Resize the image using the scaling factor
new_width = int(image.width * scaling_factor)
new_height = int(image.height * scaling_factor)
image = image.resize((new_width, new_height))

image_data = np.array(image)
print(f"image_data.shape: {image_data.shape}")
reshaped_data = image_data.reshape(-1, 3)  # Reshape to 2D array
print(f"reshaped_data.shape: {reshaped_data.shape}")
# Define different values of k

# Display images for different k values
display(HTML('<h2>Original Image</h2>'))
display(image)

k_values = [2, 5, 10, 20]  # Change these values as per your requirement

for k in k_values:
    # Apply k-means clustering
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(reshaped_data)

    # Get the labels and cluster centroids
    labels = kmeans.labels_
    centroids = kmeans.cluster_centers_

    print(image_data.shape)
    # Replace pixel values with centroid values
    new_image_data = centroids[labels].reshape(image_data.shape)

    # Convert the data back to image format
    new_image = Image.fromarray(new_image_data.astype(np.uint8))
    new_image = new_image.resize((new_width, new_height))  # Resize the clustered image

    print(f"labels.shape: {labels.shape}")
    print(f"centroids.shape: {centroids.shape}")
    # Display the clustered image
    display(HTML(f'<h2>Clustered Image (k={k})</h2>'))
    display(new_image)
```
